## Minute to hour conversion example 9

# Convert Minutes to Hours
def convert_minutes_to_hours(minutes):
    hours = minutes / 60
    return hours

# Example Usage
minutes_input = 150
hours_output = convert_minutes_to_hours(minutes_input)
print(f"{minutes_input} minutes is equal to {hours_output:.2f} hours.")

In this code cell, we have a function convert_minutes_to_hours that takes the number of minutes as input and returns the equivalent value in hours. The function divides the number of minutes by 60 (since there are 60 minutes in one hour) to calculate the hours.

In the example usage, we provide minutes_input as 150, and the function will calculate that 150 minutes is equal to 2.5 hours. The result will be printed as "150 minutes is equal to 2.50 hours."
